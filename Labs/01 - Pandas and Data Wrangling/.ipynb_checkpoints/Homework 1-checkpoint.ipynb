4) Here is the proportion of surviving passengers for each travel class

```python
#use columns age died persone
titanic.age = 1-titanic.survived
#take only columns that are needed
survived_data = titanic[['pclass','survived','age']]
#sum over the travel class
survived_plot = survived_data.groupby('pclass').sum().transpose()
#change label of 
survived_plot.index = ['survived','died']
#rename columns for convenience
survived_plot= survived_plot.rename(columns={1:'1st Class',2:'2nd Class',3:'3rd Class'})
survived_plot.plot.pie(subplots=True,autopct='%.2f',figsize=(12,4))
```

5) Here are the proportion of people that survived by travel class and sex

```python
#take only needed columns
survived = titanic[['pclass','sex','survived']]
#take only inforamtion of surviving passengers
survived = survived[survived.survived==1]
#drop useless column 'survived'
survived = survived[['pclass','sex']]

#dict to better name travel class
labec_func = {1:'1st ',2:'2nd ',3:'3rd '}

#create a new columns using travel class and sex
survived['class'] = survived['pclass'].apply(lambda x : labec_func[x] )+ survived['sex']

#use as counter
survived['pclass'] =1

#rename for more clarity
survived = survived.rename(columns={'pclass':'survived'})

#groupe by the new column 'travel class and sex'
survived = survived.groupby('class').sum().sort_values('survived')

#plot histogram
survived.plot.bar()
```
